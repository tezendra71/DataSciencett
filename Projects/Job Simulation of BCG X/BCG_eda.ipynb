```python
client_df = pd.read_csv(r'E:\DS\tezendra\Projects\Job Simulation of BCG X\client_data (1).csv')
price_df = pd.read_csv(r'E:\DS\tezendra\Projects\Job Simulation of BCG X\price_data (1).csv')
```

```python
final_df = pd.merge(client_df, price_df, on='id', how='inner')
```

```python
final_df
```

```python
# Calculate churn percentage
churn_percentage = (
    final_df["churn"]
    .value_counts(normalize=True)
    .sort_index() * 100
)

# Convert to DataFrame for stacked bar plot
churn_percentage = churn_percentage.to_frame().T
churn_percentage.index = ["Companies"]
churn_percentage.columns = ["Retention", "Churn"]

# Plot
ax = churn_percentage.plot(
    kind="bar",
    stacked=True,
    figsize=(5,5),
    color=["skyblue", "salmon"]
)

# Add percentage labels
for container in ax.containers:
    ax.bar_label(container, fmt="%.1f%%")

plt.title("Churning Status")
plt.ylabel("Percentage")
plt.xticks(rotation=0)
plt.legend(title="")
plt.show()
```
