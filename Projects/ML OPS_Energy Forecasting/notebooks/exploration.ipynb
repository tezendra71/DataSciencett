# Energy Forecasting - Exploratory Data Analysis (EDA)

In this notebook, we will explore the energy forecasting datasets (`df_train.csv` and `df_test.csv`) to understand their structure, patterns, and prepare for modeling.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
from scipy import stats
import os

data_dir = '../data'

# Load datasets
train = pd.read_csv(os.path.join(data_dir, 'df_train.csv'))
test = pd.read_csv(os.path.join(data_dir, 'df_test.csv'))

print("Train shape:", train.shape)
print("Test shape:", test.shape)

# Initial Data Inspection

display(train.head())
display(test.head())

print("\nTrain Info:")
train.info()
print("\nTest Info:")
test.info()

display(train.describe(include='all'))

# Missing values per column
print("Train missing:\n", train.isnull().sum())
print("\nTest missing:\n", test.isnull().sum())

# Visualize missingness
msno.matrix(train, figsize=(12,4))
plt.title('Missingness Matrix - Train')
plt.show()

msno.matrix(test, figsize=(12,4))
plt.title('Missingness Matrix - Test')
plt.show()

# Check duplicates
print("Train duplicates:", train.duplicated().sum())
print("Test duplicates:", test.duplicated().sum())

# Target Variable Analysis

# Distribution of power consumption
fig, axes = plt.subplots(1, 2, figsize=(14,5))
sns.histplot(train['power_consumption'], kde=True, ax=axes[0])
axes[0].set_title('Train - Power Consumption Distribution')
sns.histplot(test['power_consumption'], kde=True, ax=axes[1])
axes[1].set_title('Test - Power Consumption Distribution')
plt.tight_layout()
plt.show()


# Statistical summary of target
print("Train target stats:\n", train['power_consumption'].describe())
print("\nTest target stats:\n", test['power_consumption'].describe())

# Check for outliers using IQR
Q1 = train['power_consumption'].quantile(0.25)
Q3 = train['power_consumption'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
outliers_train = train[(train['power_consumption'] < lower_bound) | (train['power_consumption'] > upper_bound)]
print(f"Number of outliers in train: {len(outliers_train)} ({len(outliers_train)/len(train)*100:.2f}%)")


# Time Series Patterns

# Convert date to datetime for plotting (allowing pandas to infer the date format to be robust)
train['date'] = pd.to_datetime(train['date'])
test['date'] = pd.to_datetime(test['date'])

# Plot power consumption over time
plt.figure(figsize=(14,6))
plt.plot(train['date'], train['power_consumption'], label='Train', alpha=0.7)
plt.plot(test['date'], test['power_consumption'], label='Test', alpha=0.7)
plt.title('Power Consumption Over Time')
plt.xlabel('Date')
plt.ylabel('Power Consumption')
plt.legend()
plt.grid(True)
plt.show()


# Extract time components for deeper analysis
train['year'] = train['date'].dt.year
train['month'] = train['date'].dt.month
train['day'] = train['date'].dt.day
train['weekday'] = train['date'].dt.day_name()

test['year'] = test['date'].dt.year
test['month'] = test['date'].dt.month
test['day'] = test['date'].dt.day
test['weekday'] = test['date'].dt.day_name()

# Monthly Patterns

monthly_avg = train.groupby('month')['power_consumption'].mean().reset_index()
plt.figure(figsize=(10,5))
sns.barplot(data=monthly_avg, x='month', y='power_consumption', palette='viridis')
plt.title('Average Power Consumption by Month (Train)')
plt.xlabel('Month')
plt.ylabel('Average Power Consumption')
plt.show()

# Weekday Patterns

weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_avg = train.groupby('weekday')['power_consumption'].mean().reindex(weekday_order).reset_index()
plt.figure(figsize=(10,5))
sns.barplot(data=weekday_avg, x='weekday', y='power_consumption', palette='coolwarm')
plt.title('Average Power Consumption by Weekday (Train)')
plt.xlabel('Weekday')
plt.ylabel('Average Power Consumption')
plt.xticks(rotation=45)
plt.show()


# Yearly Trends (if multiple years)

if train['year'].nunique() > 1:
    yearly_avg = train.groupby('year')['power_consumption'].mean().reset_index()
    plt.figure(figsize=(8,5))
    sns.lineplot(data=yearly_avg, x='year', y='power_consumption', marker='o')
    plt.title('Average Power Consumption by Year')
    plt.xlabel('Year')
    plt.ylabel('Average Power Consumption')
    plt.show()

# Feature Analysis

feature_cols = [col for col in train.columns if col not in ['date', 'power_consumption']]
print("Available features:", feature_cols)

numeric_feats = train[feature_cols].select_dtypes(include=np.number).columns.tolist()
categorical_feats = train[feature_cols].select_dtypes(include='object').columns.tolist()

print("Numeric features:", numeric_feats)
print("Categorical features:", categorical_feats)

# Numeric Feature Distributions

if numeric_feats:
    fig, axes = plt.subplots(len(numeric_feats), 2, figsize=(14, 4*len(numeric_feats)))
    if len(numeric_feats) == 1:
        axes = [axes]
    for i, col in enumerate(numeric_feats):
        sns.histplot(train[col], kde=True, ax=axes[i][0])
        axes[i][0].set_title(f'Train - {col} Distribution')
        sns.histplot(test[col], kde=True, ax=axes[i][1])
        axes[i][1].set_title(f'Test - {col} Distribution')
    plt.tight_layout()
    plt.show()
else:
    print("No numeric features besides date/target.")


# Correlation with target (for numeric features)
if numeric_feats:
    corr = train[numeric_feats + ['power_consumption']].corr()
    plt.figure(figsize=(10,8))
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Matrix')
    plt.show()
else:
    print("No numeric features to correlate.")

# Categorical Feature Analysis

if categorical_feats:
    for col in categorical_feats:
        plt.figure(figsize=(10,4))
        sns.countplot(data=train, x=col, palette='Set2')
        plt.title(f'Train - {col} Distribution')
        plt.xticks(rotation=45)
        plt.show()
else:
    print("No categorical features.")


# Boxplots of target by categorical features (if any)
if categorical_feats:
    for col in categorical_feats:
        plt.figure(figsize=(12,5))
        sns.boxplot(data=train, x=col, y='power_consumption', palette='Set3')
        plt.title(f'Power Consumption by {col}')
        plt.xticks(rotation=45)
        plt.show()

# Final overview: compare train and test distributions for key features

if numeric_feats:
    for col in numeric_feats:
        plt.figure(figsize=(10,4))
        sns.kdeplot(train[col], label='Train', fill=True)
        sns.kdeplot(test[col], label='Test', fill=True)
        plt.title(f'Distribution Comparison: {col}')
        plt.legend()
        plt.show()

# Conclusion

print("EDA completed: Insights gathered for time-series forecasting.")
```
